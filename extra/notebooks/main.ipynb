1. Introducción: La Realidad de la Salud en CDMX
El escenario: El acceso a la salud en la capital no es una red uniforme, sino un sistema de "archipiélagos".

Carencia en aumento: Según datos de INEGI (SIDS 2024-2026), el porcentaje de población con carencia por acceso a servicios de salud en México saltó del 15.6% en 2016 al 34.2% en 2024. Esto significa que más de un tercio de la población depende críticamente de la eficiencia de la infraestructura pública.

Centralismo Geográfico: Existe una brecha brutal: el 57.29% de las unidades médicas se concentran en el área central de la CDMX, mientras que solo el 3.23% se ubica en la periferia.

Dependencia del Transporte: Para el 80% de los usuarios del sistema público (IMSS-Bienestar/SEDESA), el trayecto al hospital no es opcional ni se hace en auto privado; se recorre a pie desde estaciones de Metro, Metrobús o RTP.

2. Motivación: El "Impuesto de Riesgo"
El porqué: Porque en la CDMX, tu código postal determina qué tan probable es que seas asaltado yendo a una consulta.

El Crimen de la "Última Milla": Datos de la FGJ (2025) indican que aproximadamente el 21% de los delitos de alto impacto en la ciudad ocurren en la calle o en el transporte público. El "robo a transeúnte" sigue siendo uno de los delitos con mayor cifra negra y percepción de inseguridad.

Asignación Obligatoria: A diferencia de un servicio privado, el sistema público te asigna una clínica por jurisdicción. Si esa ruta obligatoria carece de luminarias o tiene alta incidencia delictiva, el Estado está imponiendo un riesgo físico como precondición para recibir atención médica.

3. ¿Por qué no se ha hecho? (El "Gap")
Existen tres barreras que tu equipo está rompiendo:

Silos de Datos: El portal de datos abiertos tiene los archivos, pero la Agencia Digital (ADIP), la Secretaría de Salud y la SSC no cruzan sus bases. Salud ve pacientes, Seguridad ve carpetas, pero nadie ve "el trayecto del paciente".

Sesgo del Ruteador Comercial: Google Maps y Waze optimizan para vehículos particulares y tiempo. Ignoran variables humanas como "densidad de luminarias" o "proximidad a cámaras del C5", que son vitales para alguien caminando hacia un hospital en el turno nocturno.

Visibilidad Política: Mapear la inseguridad en rutas de salud obligatorias es incómodo para el gobierno, ya que evidencia fallos en servicios básicos (luz, vigilancia) en las zonas más vulnerables.

4. Beneficios: ¿Para qué sirve saber esto?
Saber el índice de justicia espacial permite:

Priorización de Infraestructura: En lugar de poner luces "donde se ve feo", el gobierno puede instalarlas en los "Pasillos de Salud" con peor score de seguridad.

Rediseño de Paradas (GTFS): Si un hospital tiene un índice de riesgo alto desde la parada de Metro, la SEMOVI puede justificar la creación de una parada de RTP o Metrobús exactamente en la puerta del hospital.

Auditoría Ciudadana: Empodera a los comités vecinales para exigir "Senderos Seguros" con datos matemáticos, no solo con quejas subjetivas.

Optimización de Vigilancia: El C5 puede priorizar el monitoreo de cámaras en los horarios de mayor afluencia de consultas en las clínicas más vulnerables.

algunas paradas tienen coordenadas distintas pero estan a menos de 100m por lo tanto representan el mismo lugar fisico

```python
from sklearn.neighbors import BallTree
from sklearn.metrics.pairwise import haversine_distances

df = pd.read_csv("BD_limpio\stops_limpios.csv")
df.columns = [col.strip().lower() for col in df.columns]

df['agencia'] = df['stop_id'].astype(str).str[:2]

coords_rad = np.radians(df[['stop_lat', 'stop_lon']].to_numpy())
df['coords_rad'] = list(coords_rad)  

# agrupar
grupo_id = 0
df['grupo_stop_id'] = -1  
asignados = set()

# parametros
radio_metros = 100
radio_radianes = radio_metros / 6371000.0  # tierra en m

for agencia, grupo_agencia in df.groupby('agencia'):
    coords = np.stack(grupo_agencia['coords_rad'].values)
    indices = grupo_agencia.index.to_list()

    tree = BallTree(coords, metric='haversine')
    vecinos = tree.query_radius(coords, r=radio_radianes)

    for i, lista_vecinos in enumerate(vecinos):
        i_global = indices[i]
        if i_global in asignados:
            continue
        grupo = [indices[j] for j in lista_vecinos if indices[j] not in asignados]
        if grupo:
            for idx in grupo:
                df.at[idx, 'grupo_stop_id'] = grupo_id
                asignados.add(idx)
            grupo_id += 1

equivalencias = df[['stop_id', 'stop_name', 'grupo_stop_id']].sort_values('grupo_stop_id')
equivalencias.to_csv("equivalencias_parada_id.csv", index=False, encoding='utf-8')

df.drop(columns=['coords_rad'], inplace=True)
df.to_csv("BD_limpio\paradas_agrupadas.csv", index=False, encoding='utf-8')

print("agrupacion completada")
```

```python
import pdfplumber
import pandas as pd
import geopandas as gpd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

unidades = []
with pdfplumber.open("Directorio-Hospitales-y-Centros-Salud.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        for linea in text.split('\n'):
            if "CENTRO DE SALUD" in linea or "HOSPITAL" in linea:
                unidades.append({"nombre": linea, "raw": text})

df = pd.DataFrame(unidades)

locator = Nominatim(user_agent="myGeocoder")
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)

df['address'] = df['nombre'] + ", Ciudad de México" 
df['location'] = df['address'].apply(geocode)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)

df[['lat', 'lon', 'alt']] = pd.DataFrame(df['point'].tolist(), index=df.index)
gdf_salud = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat))
gdf_salud.crs = "EPSG:4326" 
```
